# Data Transforming

```python
# EXTRACTING THE YEAR
year = df_sales_orders[{'CREATEDAT', 'SALESORDERID'}]

year['CREATEDAT'] = year.CREATEDAT.astype('str')
for index_label, row_series in year.iterrows():
    string = row_series['CREATEDAT']   
    year.at[index_label , 'CREATEDAT'] = string[0:4]

year= year.rename(columns={'CREATEDAT': 'YEAR'})

# EXTRACTING THE MONTH
month = df_sales_orders[{'CREATEDAT', 'SALESORDERID'}]

month['CREATEDAT'] = month.CREATEDAT.astype('str')
for index_label, row_series in month.iterrows():
    string = row_series['CREATEDAT']   
    month.at[index_label , 'CREATEDAT'] = string[4:6]

month= month.rename(columns={'CREATEDAT': 'MONTH'})

# EXTRACTING THE DAY
day = df_sales_orders[{'CREATEDAT', 'SALESORDERID'}]

day['CREATEDAT'] = day.CREATEDAT.astype('str')
for index_label, row_series in day.iterrows():
    string = row_series['CREATEDAT']   
    day.at[index_label , 'CREATEDAT'] = string[6:8]

day= day.rename(columns={'CREATEDAT': 'DAY'})
```

```python
date = pd.merge(year, month, on=['SALESORDERID'], how='left')
date.head(3)
```

```python
date = pd.merge(date, day, on=['SALESORDERID'], how='left')
date.head(3)
```

```python
date['DATE'] = date["DAY"] + "/" + date["MONTH"] + "/" + date["YEAR"]
date['DATE'] = pd.to_datetime(date['DATE'])
date.head(3)
```

```python
# Putting the new columns "date" inside the dataset df_sales_orders
df_sales_orders = pd.merge(df_sales_orders, date, on=['SALESORDERID'], how='left')
df_sales_orders.head(1)
```

```python
query_business_partners = df_business_partner[{'PARTNERID', 
                                                  'ADDRESSID', 
                                                  'COMPANYNAME'}]
query_business_partners = query_business_partners.rename(columns={'COMPANYNAME': 'PARTNERNAME'})

query_adress = df_adress[{'ADDRESSID', 
                             'CITY', 
                             'COUNTRY', 
                             'REGION', 
                             'LATITUDE', 
                             'LONGITUDE'}]

query_product_categories = df_product_categories_text[{'PRODCATEGORYID', 
                                                          'SHORT_DESCR'}]
query_product_categories= query_product_categories.rename(columns={'SHORT_DESCR': 'CATEGORYNAME'})

query_products = df_products[{'PRODUCTID', 
                                 'PRODCATEGORYID', 
                                 'SUPPLIER_PARTNERID', 
                                 'PRICE'}]

query_products_text = df_products_text[{'PRODUCTID', 
                                           'SHORT_DESCR'}]
query_products_text= query_products_text.rename(columns={'SHORT_DESCR': 'PRODUCTNAME'})

query_sales_order_items = df_sales_order_items[{'SALESORDERID',
                                                  'SALESORDERITEM',
                                                  'PRODUCTID',
                                                  'CURRENCY',
                                                  'GROSSAMOUNT',
                                                  'NETAMOUNT',
                                                  'TAXAMOUNT',
                                                  'QUANTITY',
                                                  'QUANTITYUNIT'}]

query_sales_orders = df_sales_orders[{'SALESORDERID',
                                      'DATE',
                                         'CREATEDBY',
                                        'PARTNERID',
                                         'SALESORG',
                                        'GROSSAMOUNT',
                                        'NETAMOUNT',
                                        'TAXAMOUNT',
                                        'DELIVERYSTATUS'}]
```

# Building the Final Dataset

```python
query_sales_orders= query_sales_orders.rename(columns={'GROSSAMOUNT': 'SALES_GROSSAMOUNT'})
query_sales_orders= query_sales_orders.rename(columns={'NETAMOUNT': 'SALES_NETAMOUNT'})
query_sales_orders= query_sales_orders.rename(columns={'TAXAMOUNT': 'SALES_TAXAMOUNT'})
query_sales_orders= query_sales_orders.rename(columns={'CREATEDBY': 'EMPLOYEESID'})
query_sales_order_items= query_sales_order_items.rename(columns={'GROSSAMOUNT': 'BUYING_GROSSAMOUNT'})
query_sales_order_items= query_sales_order_items.rename(columns={'NETAMOUNT': 'BUYING_NETAMOUNT'})
query_sales_order_items= query_sales_order_items.rename(columns={'TAXAMOUNT': 'BUYING_TAXAMOUNT'})


query_products = pd.merge(query_products, query_products_text, on=['PRODUCTID'], how='left')
query_products.head(2)
```

```python
powerbi = pd.merge(query_sales_order_items, query_sales_orders, on=['SALESORDERID'], how='left')
powerbi.head(3)
```

```python
powerbi = pd.merge(powerbi, query_products, on=['PRODUCTID'], how='left')
powerbi.head(3)
```

```python
powerbi = pd.merge(powerbi, query_product_categories, on=['PRODCATEGORYID'], how='left')
powerbi.head(5)
```

```python
powerbi = pd.merge(powerbi, query_business_partners, on=['PARTNERID'], how='left')
powerbi.head(5)
```

```python
powerbi = pd.merge(powerbi, query_adress, on=['ADDRESSID'], how='left')
powerbi.head(5)
```

```python
powerbi = pd.merge(powerbi, df_employees, on=['EMPLOYEESID'], how='left')
powerbi.head(5)
```

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.99bikes.com.au/bikes/mountain/hardtail'
site = requests.get(url).text
soup = BeautifulSoup(site, 'html.parser')
```

```python
placas = soup.find_all('div', class_=('product details product-item-details'))
placa = placas[0]
ultima_pagina = soup.find_all('span', class_='label')
marcas = placa.find('a', class_=('product-item-link')).get_text().strip()
price = placa.find('span', class_='full-price').get_text().strip()
num_price = price[1:]
gold_price = placa.find('span', class_='gold-price').get_text().strip()
```

```python
for i in range(1,int(ultima_pagina)):
    url_page = f'https://www.99bikes.com.au/bikes/mountain/hardtail?p={i}'
    site = requests.get(url_page).text
    soup = BeautifulSoup(site, 'html.parser')
    placas = soup.find_all('div', class_=('product details product-item-details'))
    
    with open ('bikes_99.csv', 'a', new_line='', encoding='UTF-8') as f:
        
        for placa in placas:
            
            marcas = placa.find('a', class_=('product-item-link')).get_text().strip()
            
            try:
                price = placa.find('span', class_='full-price').get_text().strip()
                num_price = price[1:]
            except:
                
                num_price = '0'
                
            linha = marca + ';' + num_price
            print(linha)
            f.write(linha)
```
